```bash
$ cd ../../../home/vscode  # chemin relatif
$ cd /home/vscode          # chemin absolu
```

# 4 Afficher le contenu de tous les fichiers ou dossiers présents *
```bash
$ ls -a
```
# 5 Créer un fichier vide notes.txt
```bash
$ touch ./notes.txt
```
# 6 Ajouter dans ce fichier la ligne "Linux is powerful"
```bash
$ nano ./notes.txt  # puis on tape la ligne et on sauvegarde
$ echo "Linux is powerful" > ./notes.txt  # redirection
```
La redirection avec `>>` permet d'ajouter sans écraser le contenu existant.

# 7 Renommer ce fichier sans le déplacer en notes_backup.txt
```bash
$ mv ./notes.txt ./notes_backup.txt
```
# 8 Copier ce fichier dans le dossier cours de votre workspace
```bash
$ cp ./notes_backup.txt /workspaces/TermNSI/cours/notes_backup.txt
```
